```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('fastfood.xlsx')
df.head(4)

# Crie visualizações usando o arquivo fastfood.xlsx, como:
# 1. Gráfico de barras
# 2. Histograma
# 3. Distribuição
# 4. Dispersão
```

## 1. Gráfico de barras:

```python
# 1. Gráfico de barras
# Agrupar por 'restaurant' e somar 'calories'
df_grouped = df.groupby('restaurant')['calories'].sum()

# gráfico de barras
df_grouped.plot.bar()
plt.show()
```

## 1.2 - Ordenando Restaurantes com maior valor calórico

```python
# Agrupar por 'restaurant' e somar 'calories'
df_grouped = df.groupby('restaurant')['calories'].sum()

# Ordenar 'calories' em ordem decrescente
df_grouped = df_grouped.sort_values(ascending=False)

# Selecionar os 5 primeiros registros
df_grouped = df_grouped.head(5)

# gráfico de barras
df_grouped.plot.bar()
plt.show()
```

## 2. Histograma:

```python
# 2. Histograma
plt.figure(figsize=(10,6))
plt.hist(df['calories'], bins=10, edgecolor='black')
plt.title('Histograma de Calorias')
plt.xlabel('Calorias')
plt.ylabel('Frequência')
plt.show()
```

## 3. Distribuição:

```python
# 3. Distribuição (gráfico de distribuição para 'total_fat')
plt.figure(figsize=(10,6))
plt.hist(df['calories'], bins=10, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Distribuição de Calorias')
plt.xlabel('Calorias')
plt.ylabel('Densidade')
plt.show()
```

## 4. Dispersão:

```python
# 4. Dispersão (gráfico de dispersão entre 'calories' e 'total_fat')
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='calories', y='total_fat')
plt.title('Dispersão entre Calorias e Gordura Total')
plt.show()
```
