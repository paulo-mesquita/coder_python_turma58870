```python
"""Criar a seguinte função:
   - `tabelas_bd()`: retorna um dataframe com as tabelas do banco de dados.
*A função deve internamente abrir e fechar a conexão com o banco de dados `coderhouse.db`.*
"""

import sqlite3
import pandas as pd

def tabelas_bd():
  conn = sqlite3.connect('coderhouse.db')
  query = "SELECT name FROM sqlite_master WHERE type='table';"
  schema = pd.read_sql(sql=query, con=conn)
  conn.close()
  return schema

#tabelas_bd()
```

```python
"""Criar a seguinte função:
   - `salva_bd(df, nome_tabela)`: salva o dataframe na tabela `nome_tabela`.
*A função deve internamente abrir e fechar a conexão com o banco de dados `coderhouse.db`.*
"""

import sqlite3
import pandas as pd

def salva_bd(df, nome_tabela):
  conn = sqlite3.connect('coderhouse.db')
  df.to_sql(name=nome_tabela, con=conn, if_exists='replace', index=False)
  conn.close()
  return

#fastfood = pd.read_excel('fastfood.xlsx')

#salva_bd(df=fastfood, nome_tabela='fastfood2')
```

```python
"""Criar a seguinte função:
   - `carrega_bd(nome_tabela)`: carrega a tabela `nome_tabela` em um dataframe.
*A função deve internamente abrir e fechar a conexão com o banco de dados `coderhouse.db`.*
"""

import sqlite3
import pandas as pd

def carrega_bd(nome_tabela):
  conn = sqlite3.connect('coderhouse.db')
  query = f"SELECT * FROM {nome_tabela}"
  schema = pd.read_sql(sql=query, con=conn)
  conn.close()
  return schema

#carrega_bd(nome_tabela='fastfood')
```
